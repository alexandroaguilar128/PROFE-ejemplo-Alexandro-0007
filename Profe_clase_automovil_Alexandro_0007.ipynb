#### Clase automovil con 3 atributos, funciones: leerdatos(), mostrardatos()

#### Que sea interactivo ---> los inputs
#### crear 3 objetos carro1, carro2 y carro3
#### no usar while true, if, for, listas y diccionarios

![Info](https://raw.githubusercontent.com/alexandroaguilar128/PROFE-ejemplo-Alexandro-0007/refs/heads/main/infoclaseauto.png)

```python
class Automovil:
  def __init__(self, marca, placa, precio):
    self.marca = marca
    self.placa = placa
    self.precio = precio

  def leerdatos(self):
    self.marca = input("Ingrese la Marca del automovil: ")
    self.placa = input("Ingrese la Placa del automovil: ")
    self.precio = input("Ingrese el Precio del automovil:")
    return self.marca, self.placa, self.precio
  def mostrardatos(self):
    print("Marca: ", self.marca)
    print("Placas: ", self.placa)
    print("Precio: ", self.precio)
    return self.marca, self.placa, self.precio
# fin de la clase automovil

print("Carro Numero 1")
carro1 = Automovil("", "", "") # Creando objeto carro 1
carro1.leerdatos()
carro1.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("Carro Numero 2")
carro2 = Automovil("", "", "") # Creando objeto carro 2
carro2.leerdatos()
carro2.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("Carro Numero 3")
carro3 = Automovil("", "", "") # Creamdo objeto carro 3
carro3.leerdatos()
carro3.mostrardatos()
print("*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*-*")
print("software creado por Alexandro Aguilar NL = 1")
```
